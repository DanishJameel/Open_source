# Q5: Write a Pandas program to display the first and last name, and department number for all employees whose last name is "McEwen".

```python
# Filter employees with last name "McEwen"
filtered_df = employess[employess['LAST_NAME'] == 'McEwen']

# Select the desired columns
result_df = filtered_df[['FIRST_NAME', 'LAST_NAME', 'DEPARTMENT_ID']]

# Display the result
print(result_df)
```

```python
result_df
```

# Q6: Write a Pandas program to display the first, last name, salary and department number for those employees whose first name starts with the letter 'P'.

```python
# Filter employees with first name starting with 'P'
filtered_df = employess[employess['FIRST_NAME'].str.startswith('P')]

# Select the desired columns
result_df = filtered_df[['FIRST_NAME', 'LAST_NAME', 'DEPARTMENT_ID']]

# Display the result
print(result_df)
```

```python
result_df
```

# Q7: Write a Pandas program to display the first, last name, salary and department number for those employees whose first name does not contain the letter 'M'.

```python
# Filter employees with first name not containing 'M'
filtered_df = employess[~employess['FIRST_NAME'].str.contains('M')]

# Select the desired columns
result_df = filtered_df[['FIRST_NAME', 'LAST_NAME', 'SALARY', 'DEPARTMENT_ID']]

# Display the result
print(result_df)
```

```python
result_df
```

# Q8: Write a Pandas program to display the first name, last name, salary and department number in ascending order by department number.

```python
# Sort the DataFrame by department number
sorted_df = employess.sort_values('DEPARTMENT_ID')

# Select the desired columns
result_df = sorted_df[['FIRST_NAME', 'LAST_NAME', 'SALARY', 'DEPARTMENT_ID']]

# Display the result
print(result_df)
```

```python
result_df
```

# Q9: Write a Pandas program to display the first name, last name, salary and department number in descending order by first name.

```python
# Sort the DataFrame by first name in descending order
sorted_df = employess.sort_values('FIRST_NAME', ascending=False)

# Select the desired columns
result_df = sorted_df[['FIRST_NAME', 'LAST_NAME', 'SALARY', 'DEPARTMENT_ID']]

# Display the result
print(result_df)
result_df
```

# Q10: Write a Pandas program to display the first name, last name, salary and manger id where manager ids are zero.

```python
# Filter employees with manager ID equal to zero
filtered_df = employess[employess['MANAGER_ID'] == 0]

# Select the desired columns
result_df = filtered_df[['FIRST_NAME', 'LAST_NAME', 'SALARY', 'DEPARTMENT_ID']]

# Display the result
print(result_df)
result_df
```

# Q11: Write a Pandas program to display the first name, last name, salary and manger id where manager ids are not null.

```python

# Filter employees with non-null manager ID
filtered_df = employess[employess['MANAGER_ID'].notnull()]

# Select the desired columns
result_df = filtered_df[['FIRST_NAME', 'LAST_NAME', 'SALARY', 'DEPARTMENT_ID']]

# Display the result
print(result_df)
result_df
```

 # Q12: Write a Pandas program to display the first name, last name, salary and department number for employees who work either in department 70 or 90.

```python
# Filter employees working in department 70 or 90
filtered_df = employess[employess['DEPARTMENT_ID'].isin([70, 90])]

# Select the desired columns
result_df = filtered_df[['FIRST_NAME', 'LAST_NAME', 'SALARY', 'DEPARTMENT_ID']]

# Display the result
print(result_df)
```
