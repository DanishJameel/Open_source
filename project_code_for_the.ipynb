```python

# Convert categorical variables to binary
data = pd.get_dummies(df, columns=['Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'Sex', 'Race', 'PhysicalActivity', 'Asthma', 'Diabetic'])
# Convert the target variable to binary
data['HeartDisease'] = data['HeartDisease'].map({'No': 0, 'Yes': 1})
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.drop(['HeartDisease', 'AgeCategory'], axis=1), data['HeartDisease'], test_size=0.2, random_state=42)
# Creating a dictionary of classification models
models = {'Logistic Regression': LogisticRegression(),
          'Decision Tree Classifier': DecisionTreeClassifier(),
          'Random Forest Classifier': RandomForestClassifier(),
          'Gradient Boosting Classifier': GradientBoostingClassifier(),
          'XGBoost Classifier': XGBClassifier(),
          'Support Vector Classifier': SVC(probability=True)}

# Fitting and evaluating each model
for name, model in models.items():
    model.fit(X_train, y_train)
    if hasattr(model, "predict_proba"):  # check if the model has predict_proba method
        y_pred = (model.predict_proba(X_test)[:, 1] > 0.5).astype(int)
    else:
        y_pred = (model.decision_function(X_test) > 0).astype(int)  # use decision_function for RidgeClassifier
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    print(name + ' Accuracy: {:.4f}, Precision: {:.4f}, Recall: {:.4f}, F1-Score: {:.4f}, ROC AUC Score: {:.4f}'.format(acc*100, prec*100, rec*100, f1*100, roc_auc*100))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.drop(['HeartDisease', 'AgeCategory'], axis=1), data['HeartDisease'], test_size=0.2, random_state=42)
# Creating a dictionary of classification models
models = {'Logistic Regression': LogisticRegression(),
          'Decision Tree Classifier': DecisionTreeClassifier(),
          'Random Forest Classifier': RandomForestClassifier(),
          'Gradient Boosting Classifier': GradientBoostingClassifier(),
          'XGBoost Classifier': XGBClassifier(),
          'Support Vector Classifier': SVC(probability=True)}

# Fitting and evaluating each model
for name, model in models.items():
    model.fit(X_train, y_train)
    if hasattr(model, "predict_proba"):  # check if the model has predict_proba method
        y_pred = (model.predict_proba(X_test)[:, 1] > 0.5).astype(int)
    else:
        y_pred = (model.decision_function(X_test) > 0).astype(int)  # use decision_function for RidgeClassifier
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    print(name + ' Accuracy: {:.4f}, Precision: {:.4f}, Recall: {:.4f}, F1-Score: {:.4f}, ROC AUC Score: {:.4f}'.format(acc, prec, rec, f1, roc_auc))
```

```python
import matplotlib.pyplot as plt

# Collect the performance metrics for each model
acc_scores = []
prec_scores = []
rec_scores = []
f1_scores = []
roc_auc_scores = []

for name, model in models.items():
    y_pred = model.predict(X_test)
    acc_scores.append(accuracy_score(y_test, y_pred))
    prec_scores.append(precision_score(y_test, y_pred))
    rec_scores.append(recall_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))
    roc_auc_scores.append(roc_auc_score(y_test, y_pred))

# Create a bar plot for the performance metrics
x_pos = [i for i, _ in enumerate(models.keys())]
plt.bar(x_pos, acc_scores, color='blue', alpha=0.5, label='Accuracy')
plt.bar(x_pos, prec_scores, color='red', alpha=0.5, label='Precision')
plt.bar(x_pos, rec_scores, color='green', alpha=0.5, label='Recall')
plt.bar(x_pos, f1_scores, color='orange', alpha=0.5, label='F1-Score')
plt.bar(x_pos, roc_auc_scores, color='purple', alpha=0.5, label='ROC AUC Score')
plt.xticks(x_pos, models.keys(), rotation=45)
plt.legend(loc='lower right')
plt.title("performance metrics for each model")
plt.xlabel("Models")
plt.ylabel("Score")

plt.show()
```
