##Detecting outliers

```python
#Outlier Treatment

def outlier_detect(df):
    for i in df.describe().columns:
        Q1=df.describe().at['25%',i]
        Q3=df.describe().at['75%',i]
        IQR=Q3 - Q1
        LTV=Q1 - 1.5 * IQR
        UTV=Q3 + 1.5 * IQR
        x=np.array(df[i])
        p=[]
        for j in x:
            if j < LTV or j>UTV:
                p.append(df[i].median())
            else:
                p.append(j)
        df[i]=p
    return df

df1 = outlier_detect(fifa)
```

```python
#Ridge Regression
ridge = Ridge(alpha = 0.5)
%time ridge.fit(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
#Ridge Regression
ridge = Ridge(alpha = 0.5)
%time ridge.fit(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
#prediction
y_pred = ridge.predict(X_test)
print(y_pred)
```

```python
#check MAE, MSE and RMSE
print('Mean Absolute Error :', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error :', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error :', math.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#visualize model
x = y_test
y = y_pred

plt.title('Ridge Regression Model', fontsize = 15, color = 'g', pad = 12)
plt.plot(x, y, 'o', color = 'r')

m, b = np.polyfit(x, y, 1)
plt.plot(x, m * x + b, color = 'darkblue')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# Assuming you have your FIFA summary data in a DataFrame named FIFA_summary
df = FIFA_summary

# Select the relevant features and target variable
features = ['HOST', 'CHAMPION', 'RUNNER UP', 'THIRD PLACE']
target = 'Target'

# Prepare the data
X = df[features]
y = df[target]

# Convert categorical variables to numerical using one-hot encoding
X_encoded = pd.get_dummies(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Train the decision tree classifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Store multiple instances of unseen data in an array
unseen_data = [
    ['France', 'Brazil', 'Germany', 'Argentina'],
    ['Italy', 'Spain', 'Netherlands', 'England'],
    ['Germany', 'Argentina', 'Brazil', 'Italy']
]

# Convert unseen data to encoded form
X_unseen = pd.DataFrame(unseen_data, columns=features)
X_unseen_encoded = pd.get_dummies(X_unseen)

# Align feature names with the training data
X_unseen_encoded = X_unseen_encoded.reindex(columns=X_train.columns, fill_value=0)

# Make predictions on unseen data
y_pred = model.predict(X_unseen_encoded)

# Print the predicted values
print("Predicted targets for unseen data:")
print(y_pred)

# Define the category labels
category_labels = {0: 'No', 1: 'Yes'}

# Map the predicted numerical results to categories
y_pred_categorical = [category_labels[prediction] for prediction in y_pred]

# Print the predicted categories
print("Predicted categories for unseen data:")
print(y_pred_categorical)
```

```python
import pandas as pd
import jupyter_dash
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objects as go
from dash.dependencies import Input, Output

# Assuming you have your FIFA summary data in a DataFrame named FIFA_summary
df = FIFA_summary

app = jupyter_dash.JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Data Exploration Dashboard"),
    html.Div([
        html.Label("Select a year:"),
        dcc.Dropdown(
            id='year-dropdown',
            options=[{'label': str(year), 'value': year} for year in df['YEAR'].unique()],
            value=df['YEAR'].min()
        )
    ], style={'width': '300px', 'margin-bottom': '20px'}),
    html.Div(id='dashboard-output'),
    dcc.Graph(id='goals-per-team-graph')
])

@app.callback(
    [Output('dashboard-output', 'children'),
     Output('goals-per-team-graph', 'figure')],
    [Input('year-dropdown', 'value')]
)
def update_dashboard(selected_year):
    filtered_df = df[df['YEAR'] == selected_year]
    total_matches = filtered_df['MATCHES PLAYED'].sum()
    total_goals = filtered_df['GOALS SCORED'].sum()

    output = [
        html.H2(f"Results for Year {selected_year}"),
        html.Div(f"Total Matches: {total_matches}"),
        html.Div(f"Total Goals: {total_goals}")
    ]

    # Visualization: Goals scored per team
    goals_per_team = filtered_df.groupby(['CHAMPION'])['GOALS SCORED'].sum().reset_index()
    fig = go.Figure(data=[go.Bar(x=goals_per_team['CHAMPION'], y=goals_per_team['GOALS SCORED'])])
    fig.update_layout(
        title=f"Goals Scored per Team - Year {selected_year}",
        xaxis_title="Team",
        yaxis_title="Goals Scored"
    )

    return output, fig

if __name__ == '__main__':
    app.run_server(mode='inline')
```
